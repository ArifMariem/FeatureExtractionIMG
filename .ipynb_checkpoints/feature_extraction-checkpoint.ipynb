## Edge cutting


```python
canny = cv2.Canny(test_image, 30, 300)
(cnts, _) = cv2.findContours(canny , cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
coins = test_image.copy()
```

```python
cv2.drawContours(canny, cnts, -1, (255, 0, 0), 2)
```

```python
mx = (0,0,0,0)      # biggest bounding box so far
mx_area = 0
for cnt in cnts:
    x,y,w,h = cv2.boundingRect(cnt)
    cv2.rectangle(coins,(x,y),(x+w,y+h),(0,255,0),2)
    area = w*h
    if area > mx_area:
        mx = x,y,w,h
        mx_area = area

plt.imshow(coins)
```

```python
roi=test_image[mx[1]:mx[1]+mx[3],mx[0]:mx[0]+mx[2]]
plt.imshow(roi, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)

plt.show()
```

## Feature matching

feature matching fait référence à la recherche de caractéristiques correspondantes à partir de deux images similaires sur la base d'un algorithme de distance de recherche. utilisée pour trouver ou dériver et transférer des attributs de la source à l'image cible. 
Le processus de mise en correspondance des caractéristiques analyse généralement la topologie des images source et cible, détecte les modèles de caractéristiques, fait correspondre les modèles et fait correspondre les caractéristiques dans les modèles découverts.

il faut redémarrer le kernel

```python
orb= cv2.SIFT_create()

kp1,des1=orb.detectAndCompute(img1 , None)
kp2,des2=orb.detectAndCompute(img2 , None)
```

```python
bf=cv2.BFMatcher()
matches=bf.knnMatch(des1 , des2 , k=2)
good=[]
good_f=[]
for m,n in matches:
  if m.distance<0.75*n.distance:
    good.append([m])
    good_f.append(m)
img3=cv2.drawMatchesKnn(img1, kp1, img2 , kp2 , good[:100] , None , flags=2)
fig= plt.figure(figsize=(15,15))
ax=fig.add_subplot(111)
ax.imshow(img3)
plt.show()
```

 SIFT (Scale Invariant Feature Transform). Cette méthode détecte des caractéristiques invariantes à l'échelle et à la rotation de l'image. De plus, les caractéristiques sont locales et basées sur l'apparence de l'objet en certains points intéressants. Ils sont également robustes aux changements d'éclairage, au bruit et aux changements mineurs de point de vue.


##Correction rotation

```python
if len(good_f) > 100:
        src_points = np.float32([kp1[m.queryIdx].pt for m in good_f]).reshape(-1, 1, 2)
        dst_points = np.float32([kp2[m.trainIdx].pt for m in good_f]).reshape(-1, 1, 2)
        m, mask = cv2.findHomography(src_points, dst_points, cv2.RANSAC, 5.0)
        corrected_img = cv2.warpPerspective(img1, m, (img2.shape[1], img2.shape[0]))

plt.imshow(corrected_img)
```
